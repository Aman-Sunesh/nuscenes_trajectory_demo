```python
%matplotlib inline

import os
import sys
import numpy as np
from nuscenes.nuscenes import NuScenes

PROJECT_ROOT = os.getcwd()

if os.path.basename(PROJECT_ROOT) == "notebooks":
    PROJECT_ROOT = os.path.abspath(os.path.join(PROJECT_ROOT, ".."))

sys.path.append(PROJECT_ROOT)

DATA_PATH = os.path.join(PROJECT_ROOT, "data", "nuscenes_mini_traj_only.npz")

data = np.load(DATA_PATH, allow_pickle=True)

print(data.files)

for key in data.files:
    print(key, data[key].shape)

# Raw nuScenes folder, NOT the .npz file
NUSCENES_ROOT = os.path.join(PROJECT_ROOT, "data", "sets", "nuscenes")

nusc = NuScenes(
    version="v1.0-mini",
    dataroot=NUSCENES_ROOT,
    verbose=True
)
```
